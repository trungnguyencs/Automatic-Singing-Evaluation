```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.redkaraoke.com/profile/maaikehuisman1/recordings/1155684")
```

```python
soup = BeautifulSoup(r.content, "lxml")
```

```python
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
for link in soup.find_all('a'):
    print(link.text, link.get('href'))
```

```python
for link in soup.find_all('a'):
   "<a href='%s'>%s</a>" %(link.get('href'),link.text)
```
