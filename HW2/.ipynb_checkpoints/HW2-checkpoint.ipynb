Heat map for the filled data frame

```python
df_dropped = df.drop(["hashottuborspa", "propertycountylandusecode", "propertyzoningdesc", 
                            "fireplaceflag", "taxdelinquencyflag", "assessmentyear", 
                            "airconditioningtypeid", "buildingclasstypeid", "decktypeid",
                            "heatingorsystemtypeid", "regionidcity", "regionidcounty",
                            "poolcnt", "fips", "buildingqualitytypeid", "threequarterbathnbr",
                            "regionidneighborhood", "regionidzip", "storytypeid",
                            "pooltypeid7", "rawcensustractandblock", "censustractandblock",
                            "typeconstructiontypeid", "pooltypeid10", "pooltypeid2", "architecturalstyletypeid"], axis=1)
df_dropped_filled = df_dropped.fillna(0)

corr = df_dropped_filled.corr(method='pearson', min_periods=1)
fig1, ax1 = plt.subplots(figsize=(18,18))  
sns.heatmap(corr, square=True, ax=ax1, linewidths=0.1, annot=True)
plt.show()
```

Finding the most negative correlation (correlation value is closest to -1)

```python
corr_min = corr.min(axis=0)
corr_min = corr_min.min()

most_negative1 = 'parcelid'
most_negative2 = 'parcelid'
for index1 ,row in corr.iterrows():
    for index2, cell in row.iteritems():
        if cell == corr_min:
            most_negative1 = index1
            most_negative2 = index2
            break
print('The most negative correlation: %s - %s') %(most_negative1, most_negative2)
```

Finding the most positive correlation (correlation value is closest to 1, except values in the diagonal line)

```python
max_value=-2
max_x=None
max_y=None
for index1, row in corr.iterrows():
    for index2, item in row.iteritems():
        if index1==index2:
            continue
        if item>max_value:
            max_value=item
            max_x=index1
            max_y=index2

print('The most positive correlation: %s - %s') %(max_x, max_y)
```

## 1. A description of how it works:

(I don't really understand this question so I am just listing both the theoretical knowledge and details of what I have done for this model)

### Theoretical knowledge:
The linear model can be presented as: 

$y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n$

- $y$ is the response
- $\beta_0$ is the intercept
- $\beta_1$ is the coefficient for $x_1$ (the first feature)
- $\beta_n$ is the coefficient for $x_n$ (the nth feature)

The $\beta$ values are called the **model coefficients**. These values are "learned" during the model fitting step using the "least squares" criterion. Then, the fitted model can be used to make predictions.

How this models work: Basically it is the problem of finding the best-fitting straight line through the set of points (Finding the regression line, or finding the values of set of features $\beta$) 

It can be translated to minimizing one of the following metrics:

**Mean Squared Error** (MSE) is the mean of the squared errors:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

- **MAE** : the average error.
- **MSE** : "punishes" larger errors, sensitive with outliers.
- **RMSE**: nearly the same meaning with MSE, but RMSE is interpretable in the "y" units.
